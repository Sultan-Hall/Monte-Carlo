# Estimating Pi using Monte Carlo Methods: The Inscribed Circle Method

### The purpose of this project is to estimate pi, the ratio of a circle's circumference to its diameter, by approximating it. 
- This can be done by simply taking the unit square, and inscribing one fourth of a circle in it. 
- We then randomly add points to the area inside the inscribed rectangle, and calculate how many fall within the circle and how many total points there are.
- We then get the ratio of how many fall within to how many total points there are
- We multiply this total by 4 in order to represent the entirety of a circle, approximating pi.
    - We can use only one fourth of the circle to improve run-time due to the graphics generation.
- To generate a graphic along with estimating pi, matplotlib and numpy were used. Matplotlib produced the graphic, while numpy was used to create arrays of data that come from running the program with different inputs (differing numbers of points).
- Because the method is based on a graphic plot, rather than simply calculating the distance to the edge from each point, using a large number of points makes the program take a long time to run, so the largest number of points used is 15000.



```python

import random

import numpy
from matplotlib import*
from numpy import*
import numpy as np
from math import*
import random
from random import *
import matplotlib.pyplot as plt


def carlo_Dot(n):
    x_Coord= 0
    y_Coord = 0
    center = x_Coord, y_Coord

    # Unit circle
    circle = plt.Circle(center, radius=1)
    plt.gca().add_patch(circle)
    transformedC = circle.get_transform().transform((1,1))

    # Unit Square
    square = plt.Rectangle((x_Coord, y_Coord), 1, 1, edgecolor="red", facecolor="None")
    plt.gca().add_patch(square)
    transformedS = square.get_transform().transform((.5,.5))



    plt.xlim(0, 5)
    plt.ylim(0, 5)

    # Lists, only used for their size and append properties.
    square_list = []
    circle_list = []

    for i in (range(n)):

        # Creates pseudorandom points within the square
        rand_x = uniform(x_Coord, x_Coord +1)
        rand_y = uniform(y_Coord, y_Coord + 1)

        # Creates a tuple
        tuple = (rand_x, rand_y)

        #  transforms points because matplotlib is matplotlib
        transformedC = circle.get_transform().transform(tuple)
        transformedS = square.get_transform().transform(tuple)

        # adds points to plot
        point = plt.Circle(tuple, radius=.01, color = "green")
        plt.gca().add_patch(point)

        # Checks if the point is in the square, and then if it's in the circle

        if square.contains_point(transformedS):
            square_list.append(tuple)
            if circle.contains_point(transformedC):
                circle_list.append(tuple)

    ratio = str((len(circle_list)/len(square_list))*4)

    return ratio

# Makes the function take in an array as an input and output another array containing floats
vectorized_carlo = np.vectorize(carlo_Dot, otypes= [float])

# testing values, smaller dataset used to show random fluctuations
values = np.array([20,30,40,50,100, 300, 700])

# Applies the "vectorizing" to the list of values above
y_axis = vectorized_carlo(values)


plt.figure(0)
plt.scatter(values, y_axis)
plt.show()
```
